```python
df3_en=pd.get_dummies(df2['brand'])
df3_en
```

```python
df3=pd.concat([df2,df3_en],axis=1)
df3
```

```python
df4_en=pd.get_dummies(df3['owner'])
df4=pd.concat([df3,df4_en],axis=1)
df4
```

```python
df5=df4.drop(['bike_name' , 'city', 'brand' , 'owner'],axis=1)
df5
```

# APPLYING MODEL

```python
X=df5.iloc[: , 1:]
Y=df5.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
std_reg=LinearRegression()
#Train or fit the training data 
std_reg.fit(x_train,y_train)
```

```python
y_predict=std_reg.predict(x_test)
```

```python
slr_score=std_reg.score(x_test,y_test)
print(slr_score)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import ShuffleSplit
cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
cross_val=cross_validate(std_reg,X,Y,cv=cv)
cross_val
```

```python
import numpy as np
std_reg_avg=np.average(cross_val['test_score'])
print(std_reg_avg)
```
