```python
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/Samarth Thakur/Downloads/04+-+decisiontreeAdultIncome.csv")
data
```

```python
data.isnull().any()
```

```python
newdata=data.copy()
newdata
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
newdata['wc']=le.fit_transform(newdata['wc'])
newdata['wc'].unique()
```

```python
newdata
```

```python
#for fitting multiple column in one label encoder
cat_col=['education','marital status','race','gender']

for col in cat_col:
    newdata[col]=le.fit_transform(newdata[col])
```

```python
newdata
```

```python
X=newdata.iloc[: ,: -1]
Y=newdata.iloc[: , -1]
print(X)
print("-----------------------------------------------------------------")
print(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=\
    train_test_split(X,Y,test_size=0.3,random_state=1234)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(random_state=42)
dtc.fit(X_train,Y_train)
Y_predict=dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_predict)
score=dtc.score(X_test,Y_test)
print(score*100)
print("------------------------")
print(cm)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rtc=RandomForestClassifier(random_state=42)
rtc.fit(X_train,Y_train)
Y_predict=rtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(Y_test,Y_predict)
score1=rtc.score(X_test,Y_test)
print(score1*100)
print("------------------------")
print(cm1)
```
